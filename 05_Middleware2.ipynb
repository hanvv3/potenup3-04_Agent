## 2) Human-in-the-loop (HITL)

```python
from langchain.agents import create_agent
from langchain.agents.middleware import HumanInTheLoopMiddleware, wrap_tool_call
from langchain.tools import tool
from langgraph.checkpoint.memory import InMemorySaver

from collections.abc import Callable
from langchain.messages import ToolMessage
from langchain.tools.tool_node import ToolCallRequest
from langgraph.types import Command
```

```python
@tool
def search_docs(query: str) -> str:
    '''문서를 검색합니다.'''
    return f"[search] '{query}' 검색 결과입니다."

@tool
def delete_file(filename: str) -> str:
    '''문서를 삭제합니다.'''
    return f"[delete] '{filename}' 파일을 삭제합니다."

@wrap_tool_call
def log_tool_calls(
    request: ToolCallRequest,
    handler: Callable[[ToolCallRequest], ToolMessage | Command],
) -> ToolMessage | Command:
    tool_name = request.tool_call['name']
    tool_args = request.tool_call['args']
    
    print(f"[TOOL START]{tool_name}/args={tool_args}")
    
    try:
        result = handler(request)

        if isinstance(result, ToolMessage):
            print(f"[TOOL END]{tool_name}/result={result.content}")
        else:
            print(f"[TOOL END]{tool_name}/result={result}")

        return result

    except Exception as e:
        print(f"[TOOL ERROR]{tool_name}/error={e}")
        raise


# 2) 체크포인터 필수
checkpointer = InMemorySaver()

agent = create_agent(
    model='openai:gpt-5.4-mini',
    tools=[search_docs, delete_file],
    middleware=[
        HumanInTheLoopMiddleware(
            interrupt_on={
                'delete_file': True,
            },
            description_prefix='검토 필요 작업'
        ),
        log_tool_calls
    ],
    checkpointer=checkpointer
)

config = {
    'configurable': {
        'thread_id': 'hitl-demo-1'
    }
}
```

```python
agent.invoke(
    {
        'messages': [
            {
                'role': 'user',
                'content': "'old_report.pdf'파일을 삭제해줘"
            }
        ]
    },
    config=config,
)
```

```python
resume_result = agent.invoke(
    Command(resume={'decisions':[{'type':'approve'}]}),
    config=config,
    #version='v2'
)
```

```python
print(f"{resume_result['messages'][-1].content}")
```

### HITL 실전 예시

```python
from langchain.agents import create_agent
from langchain.agents.middleware import HumanInTheLoopMiddleware, wrap_tool_call
from langgraph.checkpoint.memory import InMemorySaver
from langgraph.types import Command
from langchain.tools import tool

from collections.abc import Callable
from langchain.messages import ToolMessage
from langchain.tools.tool_node import ToolCallRequest

from pydantic import BaseModel, Field
```

```python
# -----------------------------
# 1) 실제 실행될 민감한 툴
# -----------------------------
@tool
def delete_file(filename: str) -> str:
    '''파일을 삭제합니다.'''
    return f"[DELETE] '{filename}'을(를) 삭제하였습니다."
```

```python
# -----------------------------
# 2) 삭제 승인 여부를 판정할 구조화 출력 스키마
# -----------------------------
class ApprovalDecision(BaseModel):
    approved: bool = Field(description='사용자의 발화가 삭제 승인이면 True, 아니면 False 반환')
    reason: str = Field(description='판단이유를 짧게 설명')
    

@wrap_tool_call
def log_tool_calls(
    request: ToolCallRequest,
    handler: Callable[[ToolCallRequest], ToolMessage | Command],
) -> ToolMessage | Command:
    tool_name = request.tool_call['name']
    tool_args = request.tool_call['args']
    
    print(f"[TOOL START]{tool_name} / args={tool_args}")
    
    try:
        result = handler(request)

        if isinstance(result, ToolMessage):
            print(f"[TOOL END]{tool_name} / result={result.content}")
        else:
            print(f"[TOOL END]{tool_name} / result={result}")

        return result

    except Exception as e:
        print(f"[TOOL ERROR]{tool_name} / error={e}")
        raise
```

```python
# -----------------------------
# 3) 승인 판정 전용 에이전트
#    - tools 없음
#    - response_format으로 구조화 출력 사용
# -----------------------------
approval_agent = create_agent(
    model='openai:gpt-5.4-nano',
    tools = [],
    system_prompt=(
        '당신은 Human-In-The-Loop 파일 삭제 승인 판별기입니다. '
        '사용자는 이미 파일 삭제 요청을 했고, 현재 시스템은 삭제 실행 직전에 멈춘 상태입니다. '
        '이제 사용자의 후속 발화가 "중단된 파일 삭제 작업을 승인하는지" 판단하세요. '

        'approved=True 조건: '
        '사용자가 삭제 진행, 실행, 승인, 허용, 동의의 의미를 표현하면 True입니다. '
        "예: '승인', '승인합니다', '네', '응', '삭제해줘', '지워줘', "
        "'삭제하세요', '진행하세요', '삭제해도 됩니다', '그대로 진행', 'yes', 'ok'. "

        'approved=False 조건: '
        '취소, 거절, 보류, 질문, 불확실한 표현이면 False입니다. '
        "예: '아니', '취소', '잠깐', '삭제하지 마', '정말 삭제돼?', '무슨 파일?'. "

        '반드시 ApprovalDecision 스키마로만 답하세요.'
    ),
    response_format=ApprovalDecision,
)

# create_agent의 structured output은 structured_response에 담겨 반환됩니다.
# 직접 schema type을 넘기면 LangChain이 적절한 structured output strategy를 고릅니다.
```

```python
# -----------------------------
# 4) HITL 본체 에이전트
# -----------------------------
checkpointer = InMemorySaver()

agent = create_agent(
    model='openai:gpt-5.4-mini',
    tools=[delete_file],
    middleware=[
        log_tool_calls,
        HumanInTheLoopMiddleware(
            interrupt_on={
                'delete_file': {
                    'allowed_decisions' : ['approve', 'reject'],
                    'description': '파일 삭제는 사람이 승인을 한 후에만 진행됩니다.'
                }
            },
            description_prefix='사람의 검토가 필요한 작업'
        )
    ],
    checkpointer=checkpointer
)

config = {
    'configurable': {
        'thread_id' : 'delete_approval_llm_demo'
    }
}
```

```python
# -----------------------------
# 5) 첫 요청 처리: 삭제 요청을 받으면 interrupt 발생
# -----------------------------
def request_delete(prompt: str):
    result = agent.invoke(
        {
            'messages' : [
                {
                    'role': 'user',
                    'content': prompt
                }
            ]
        },
        config=config,
        version='v2'
    )
    
    interrupts = result.interrupts
    messages = result.value['messages']
    
    if interrupts:
        print('삭제를 하려면 승인여부를 입력해주세요. (예: 삭제해줘 등등)')
        return 'WAITING_APPROVAL'
    
    # interrupt 없이 끝난 경우
    print(f"{result['messages'][-1].content}")
    return 'DONE'


# -----------------------------
# 6) 두 번째 사용자 입력을 LLM으로 판정 후 재개
# -----------------------------
def handle_followup_with_llm(user_text: str):
    judge_result = approval_agent.invoke(
        {
            'messages': [
                {
                    'role': 'user',
                    'content': (
                        '상황: 사용자는 이미 파일 삭제를 요청했고, '
                        '시스템은 삭제 실행 직전에 사람의 승인을 기다리는 중입니다.\n'
                        '질문: 다음 후속 발화가 대기 중인 파일 삭제 작업을 승인하는 뜻입니까?\n'
                        f'후속 발화: {user_text}'
                    )
                }
            ]
        }
    )
    
    decision: ApprovalDecision = judge_result['structured_response']

    print(f"[판정]approved={decision.approved}, reason={decision.reason}")
    
    if decision.approved:
        resume_result = agent.invoke(
            Command(
                resume={
                    'decisions': [
                        {'type': 'approve'}
                    ]
                }
            ),
            config=config,
            version='v2',
        )
        print(f"{resume_result['messages'][-1].content}")
        return "APPROVED"
    else:
        resume_result = agent.invoke(
            Command(
                resume={
                    'decisions': [
                        {
                            'type': 'reject',
                            'reason': f"사용자의 승인 문구로 보기 어려움: {decision.reason}"
                        }
                    ]
                }
            ),
            config=config,
            version='v2'
        )
        print('삭제하지 않았습니다.')
        return "REJECTED"
```

```python
# -----------------------------
# 7) 실행 예시
# -----------------------------
state = request_delete("'old_report.pdf' 파일을 삭제해줘.")

if state == "WAITING_APPROVAL":
    followup = input('사용자 입력: ')
    handle_followup_with_llm(followup)
```

```python
thread_id = 'delete_approval_llm_demo'
checkpointer.delete_thread(thread_id=thread_id)
```

## 3) ToolCallLimit

```python
from langchain.agents import create_agent
from langchain.agents.middleware import ToolCallLimitMiddleware, wrap_tool_call
from langgraph.checkpoint.memory import InMemorySaver
from langchain.tools import tool
```

```python
@tool
def search_docs(query: str) -> str:
    '''문서를 검색합니다.'''
    return f"[SEARCH] '{query}' 검색 결과입니다."

@tool
def make_image(query: str) -> str:
    """이미지를 생성합니다.."""
    return f"[IMG GENERATED] '{query}' 이미지가 생성되었습니다."

@wrap_tool_call
def log_tool_calls(
    request: ToolCallRequest,
    handler: Callable[[ToolCallRequest], ToolMessage | Command],
) -> ToolMessage | Command:
    tool_name = request.tool_call['name']
    tool_args = request.tool_call['args']
    
    print(f"[TOOL START]{tool_name} / args={tool_args}")
    
    try:
        result = handler(request)

        if isinstance(result, ToolMessage):
            print(f"[TOOL END]{tool_name} / result={result.content}")
        else:
            print(f"[TOOL END]{tool_name} / result={result}")

        return result

    except Exception as e:
        print(f"[TOOL ERROR]{tool_name} / error={e}")
        raise
```

```python
checkpointer = InMemorySaver()

agent = create_agent(
    model='openai:gpt-5.4-nano',
    tools=[search_docs, make_image],
    middleware=[
        log_tool_calls,
        ToolCallLimitMiddleware(run_limit=3),
        ToolCallLimitMiddleware(
            tool_name='make_image',
            thread_limit=5,
            run_limit=1,
        )
    ],
    checkpointer=checkpointer
)

config = {
    'configurable': {'thread_id': 'tool-limit-demo-1'}
}
```

```python
result = agent.invoke(
    {
        'messages': [
            {
                'role': 'user',
                'content': "고양이 이미지를 만들어줘. 개 이미지도 만들어줘. a.pdf 문서를 검색해주고, b.pdf 문서도 검색해줘."
            }
        ]
    },
    config=config
)
```

```python
print(f"{result["messages"][-1].content}")
```

## 4) ModelFallback

```python
from langchain.agents import create_agent
from langchain.agents.middleware import ToolCallLimitMiddleware, ModelFallbackMiddleware, wrap_tool_call
from langgraph.checkpoint.memory import InMemorySaver
from langchain.tools import tool

# uv add langchain-google-vertaxai
from langchain_google_genai import ChatGoogleGenerativeAI # gpt-5.4-mini가 되지 않으면 gemini로 바꿈
```

```python
gemini_fallback = ChatGoogleGenerativeAI(
    model='gemini-3.1-flash',
    temperature=0
)
```

```python
@tool
def search_docs(query: str) -> str:
    '''문서를 검색합니다.'''
    return f"[SEARCH] '{query}' 검색 결과입니다."

@tool
def make_image(query: str) -> str:
    """이미지를 생성합니다.."""
    return f"[IMG GENERATED] '{query}' 이미지가 생성되었습니다."

@wrap_tool_call
def log_tool_calls(
    request: ToolCallRequest,
    handler: Callable[[ToolCallRequest], ToolMessage | Command],
) -> ToolMessage | Command:
    tool_name = request.tool_call['name']
    tool_args = request.tool_call['args']
    
    print(f"[TOOL START]{tool_name} / args={tool_args}")
    
    try:
        result = handler(request)

        if isinstance(result, ToolMessage):
            print(f"[TOOL END]{tool_name} / result={result.content}")
        else:
            print(f"[TOOL END]{tool_name} / result={result}")

        return result

    except Exception as e:
        print(f"[TOOL ERROR]{tool_name} / error={e}")
        raise
```

```python
checkpointer = InMemorySaver()

agent = create_agent(
    model='openai:gpt-5.4-nano',
    tools=[search_docs, make_image],
    middleware=[
        log_tool_calls,
        ToolCallLimitMiddleware(run_limit=3),
        ToolCallLimitMiddleware(
            tool_name='make_image',
            thread_limit=5,
            run_limit=1,
        ),
        ModelFallbackMiddleware(
            gemini_fallback,
            'anthropic:claude-sonnet-4-5'
        )
    ],
    checkpointer=checkpointer
)

config = {
    'configurable': {'thread_id': 'tool-limit-demo-1'}
}
```

```python
result = agent.invoke(
    {
        'messages': [
            {
                'role': 'user',
                'content': "고양이 이미지를 만들어줘. 개 이미지도 만들어줘. a.pdf 문서를 검색해주고, b.pdf 문서도 검색해줘."
            }
        ]
    },
    config=config
)
```

```python
print(f"{result["messages"][-1].content}")
```

## 5) PIIMiddleware

```python
from langchain.agents import create_agent
from langchain.agents.middleware import PIIMiddleware

agent = create_agent(
    model="openai:gpt-5.4-mini",
    tools=[],
    middleware=[
        # 이메일은 [REDACTED_EMAIL] 형태로 가림
        PIIMiddleware(
            "email",
            strategy="redact",
            apply_to_input=True,
        ),
        # 신용카드는 일부만 남기고 마스킹
        PIIMiddleware(
            "credit_card",
            strategy="mask",
            apply_to_input=True,
        ),
    ],
)

result = agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": (
                    "내 이메일은 minsu@example.com 이고, "
                    "카드번호는 5105-1051-0510-5100 이야. "
                    "이 정보로 고객 상담 메모를 만들어줘."
                ),
            }
        ]
    }
)

print(result["messages"][-1].content)
```
