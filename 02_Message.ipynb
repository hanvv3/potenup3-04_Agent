```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import SystemMessage, HumanMessage

model = init_chat_model('openai:gpt-5.4-mini')

msg = [
    SystemMessage('당신은 인공지능 에이전트입니다. 모르는건 답하지 말아주세요.'),
    HumanMessage('머신러닝이란?')
]

result = model.invoke(msg)

result
```

```python
from langchain.agents import create_agent
from langchain.tools import tool

@tool
def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"

@tool
def add_number(a: int, b: int) -> int:
    '''Sum of two integers'''
    return a + b

agent = create_agent(
    model="openai:gpt-5.4-mini",
    tools=[get_weather, add_number],
    system_prompt="You are a helpful assistant"
)

result = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": "1. 서울 날씨는 어때?, 2. 3+5+8은?"}
        ]
    }
)

result
```

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun, YouTubeSearchTool

search = DuckDuckGoSearchRun()
youtube = YouTubeSearchTool()

agent = create_agent(
    model='gpt-5.4-mini',
    tools=[search, youtube],
    system_prompt='You are a helpful assistant'    
)

result = agent.invoke(
    {
        'messages':[
            {'role': 'user', 'content': '토토로 주제가 찾아서 유튜브 주소로 알려줘.' }
        ]
    }
)
```

```python
import os
from langchain_google_community import GmailToolkit
from langchain.agents import create_agent
from langchain.tools import tool
from pydantic import BaseModel, Field

# 1. Pydantic 모델 정의
# 이 모델은 LLM이 자연어에서 어떤 데이터를 추출해야 하는지 정의하는 가이드 역할을 합니다.
class EmailStructure(BaseModel):
    emailAddrFrom: str = Field(description='보내는 사람의 이메일 주소 (Gmail API 사용 시 대개 무시됨)')
    emailAddrTo: str = Field(description='받는 사람의 이메일 주소')
    title: str = Field(description='이메일의 제목')
    content: str = Field(description='이메일의 내용')

# 2. 실제 Gmail 기능을 수행하는 커스텀 도구 정의
# GmailToolkit의 기본 기능을 사용자의 EmailStructure 스키마와 연결합니다.
toolkit = GmailToolkit()
gmail_tools = {t.name: t for t in toolkit.get_tools()}

@tool(args_schema=EmailStructure)
def send_email_via_gmail(emailAddrFrom: str, emailAddrTo: str, title: str, content: str) -> str:
    '''사용자의 Gmail 계정을 사용하여 실제로 이메일을 발송합니다.'''
    # GmailToolkit의 send_message 도구를 내부적으로 호출합니다.
    send_tool = gmail_tools["send_gmail_message"]
    result = send_tool.invoke({
        "to": emailAddrTo,
        "subject": title,
        "message": content
    })
    return f"이메일 발송 결과: {result}"

# 에이전트가 사용할 도구 리스트
tools = [send_email_via_gmail]

# 3. v1.x 표준 에이전트 생성
# 모델 식별자를 직접 사용하며, 별도의 PromptTemplate 객체 없이 system_prompt로 설정을 완료합니다.
agent = create_agent(
    model="openai:gpt-5.4-mini",  # 현재 사용 가능한 가장 최신 모델을 권장합니다.
    tools=tools,
    system_prompt=(
        "당신은 이메일 관리를 전문으로 하는 비서입니다. "
        "사용자의 요청에 따라 이메일을 작성하고 'send_email_via_gmail' 도구를 사용하여 전송하십시오."
    )
)

# 4. 에이전트 실행 및 결과 출력
# v1.x 표준: 입력을 'messages' 키의 리스트 구조로 전달합니다.
response = agent.invoke({
    "messages": [
        {
            "role": "user", 
            "content": (
                "hanvv3@gmail.com에서 hanvv3@gmail.com으로"
                "내용은 '이번 주 금요일 미팅은 예정대로 진행됩니다.'라고 작성해줘."
                "제목은 내용을 요약해서 적당한 제목으로 이메일 보내줘."
                "오류가 났다면 왜 났는지 자세하게 설명해줘."
            )
        }
    ]
})

print("\n[에이전트 작업 과정 및 결과]")
# 마지막 메시지(AI의 최종 응답)를 출력합니다.
print(response["messages"][-1].content)
```
