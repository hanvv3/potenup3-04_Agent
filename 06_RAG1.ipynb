## 2) Split

```python
# uv add langchain-text-splitters
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = '''
미국 하원 공화당 의원들이 쿠팡 등 미국 기업들에 대한 차별적인 규제를 즉각 중단하라고 서한을 보낸 데 대해 외교부는 차별은 없다고 반박했습니다.

외교부 당국자는 오늘(23일) "우리 정부는 한미 양국 정상이 '조인트 팩트시트'에서 합의한 대로 미국 디지털 기업에 대한 차별적 조치나 불필요한 장벽에 직면하게 하지 않겠다는 약속을 충실히 이행하고 있으며, 쿠팡도 이에 해당한다"고 밝혔습니다.

이어 "쿠팡에 대한 조사와 조치는 우리 국내법과 적법절차에 따라 이뤄지고 있으며, 이는 국적과 무관하게 비차별적으로 진행되는 것"이라면서 "이런 입장을 미 의회에 지속 설명해왔으며 앞으로도 정부의 기본 입장을 설명해 나갈 예정"이라고 덧붙였습니다.

앞서 미국 연방 하원 공화당 의원 모임인 '공화당 연구위원회'(RSC) 소속 의원 54명은 한국에서 사업하는 미국 기업들에 대한 차별적인 규제를 즉각 중단해달라는 내용의 서한을 강경화 주미대사에게 보냈습니다.

의원들은 쿠팡에 대해 "지난 10년간 미국의 한국 대상 외국인직접투자(FDI)의 최대 원천이었다"면서 "안타깝게도 한국 정부는 민감도가 낮은 정보 유출 사건을 구실로 쿠팡에 범정부적 공격을 가했다"고 주장했습니다.
'''

splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=30,
)

chunks = splitter.split_text(text)

print(len(chunks))
chunks
```

```python
from langchain_text_splitters import Language, RecursiveCharacterTextSplitter

code = """
def greet(name):
    return f"Hello, {name}"

def add(a, b):
    return a + b

class User:
    def __init__(self, name):
        self.name = name
"""

splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON,
    chunk_size=80,
    chunk_overlap=10
)

chunks = splitter.split_text(code)

for i, chunk in enumerate(chunks):
    print(f"[Chunk {i}]")
    print(chunk)
    print("-" * 30)
```

```python
from langchain_docling import DoclingLoader
from langchain_docling.loader import ExportType
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings

loader = DoclingLoader(
    file_path=['./data/wt_data.pdf'],
    export_type=ExportType.MARKDOWN
)

splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=100
)

embeddings = OpenAIEmbeddings()

# load
documents = loader.load()
# split
doc_chunks = splitter.split_documents(documents)
# embed
texts = [doc.page_content for doc in doc_chunks]
vectors = embeddings.embed_documents(texts)

print(len(vectors[0]))
vectors
```
